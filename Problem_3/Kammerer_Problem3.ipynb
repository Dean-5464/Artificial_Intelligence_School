```
# Predicting House Prices

# Importing the libraries
import pandas as pd #read dataset
import numpy as np #arrays and lists
import keras #build neural network
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense, Dropout #create neural network layers
from keras.models import Sequential
from keras.optimizers import Adam

# Importing the dataset
dataset = pd.read_csv('Problem3_Values.csv') #File needs to be uploaded each time you open notebook

# Getting separately the features and the target
X = dataset.iloc[:, 1:].values #removes first column (value 1)
X = X[:, np.r_[0:1,3:7]] #removes Value 4
y = dataset.iloc[:, 0].values #creates array with only value 1
#print(X[0:5]) #print first 5 rows of X
#print(y[0:5]) #print first 5 values of y - not in book (nice way to test)

# Splitting the dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 0)
 #X_Train & X_Test holds house features 85/15 split based on test_size
 #y_Train $ Y_Test holds house prices 85/15 split

 #random_state - seeds the random number generator
# Scaling the features
xscaler = MinMaxScaler(feature_range = (0,1)) #convert all values to be between 0 and 1
X_train = xscaler.fit_transform(X_train)
X_test = xscaler.transform(X_test)
#print(X[0:1]) #Real values
#print(X_train[0:1]) #Values after scaling

# Scaling the target
yscaler = MinMaxScaler(feature_range = (0,1)) #convert all values to be between 0 and 1
y_train = yscaler.fit_transform(y_train.reshape(-1,1)) #fake second dimension 1D to 2D - avoid format error
y_test = yscaler.transform(y_test.reshape(-1,1))
#print(y[0:1]) #Real values
#print(y_train[0:1]) #Values after scaling

# Building the Artificial Neural Network
model = Sequential() #create model
 #add layer to network - 64 neurons, uniform weights, relu - Rectified Linear Units, 17 - number of dimensions or columns
model.add(Dense(units = 64, kernel_initializer = 'uniform', activation = 'relu', input_dim = 5))
model.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu')) #used for output
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'relu')) #used for output
model.compile(optimizer = Adam(lr = 0.001), loss = 'mse', metrics = ['mean_absolute_error'])
 #how model is trained lr - learning rate, loss - calculate error of output

# Training the Artificial Neural Network
model.fit(X_train, y_train, batch_size = 32, epochs = 500, validation_data = (X_test, y_test))
 #batch_size - how many records before updating weights
 #epochs - how many times you train model on entire dataset
 
# Making predictions on the test set while reversing the scaling
y_test = yscaler.inverse_transform(y_test)
prediction = yscaler.inverse_transform(model.predict(X_test))

# Computing the error rate
error = (abs(prediction - y_test)/y_test) * 100
print(np.mean(error))
```
